```python
# ...existing code...
import pandas as pd

# load files
df = pd.read_csv('relationships.csv')
df_elem = pd.read_csv('elements.csv')

# map element names to element_id (fast, memory-efficient)
element_to_id = df_elem.drop_duplicates('element').set_index('element')['element_id'].to_dict()
df['element_id'] = df['element_name'].map(element_to_id)

# normalize cote and build glyph_image as "glyph_cote-codex_id.jpg"
df['glyph_cote'] = df['glyph_cote'].astype(str).str.replace('--', '-', regex=False)
df['glyph_image'] = df['glyph_cote'] + "-" + df['codex_id'].astype(str) + ".jpg"

# aggregate elements per glyph (unique, joined by |)
group_cols = ['codex_id', 'glyph_cote', 'glyph_id', 'glyph_lecture', 'glyph_image']
glyph_elements = (
    df.groupby(group_cols, dropna=False, sort=False)
      .agg(
          elements = ('element_name', lambda s: '|'.join(pd.unique(s.dropna().astype(str)))),
          element_ids = ('element_id', lambda s: '|'.join(pd.unique(s.dropna().astype(str))))
      )
      .reset_index()
)


glyph_elements.head(10)
```

```python
import os
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from typing import List, Optional, Dict

def make_multilabel_df(glyph_list: List[str],
                       glyph_key: str = 'glyph_cote',
                       glyph_elements_csv: str = 'glyph_elements.csv',
                       base_image_root: str = r'f:\CODEX\AI_clinic\glyphs',
                       glyph_folder_map: Optional[Dict[str, str]] = None,
                       use_element_ids: bool = False,
                       save_csv: Optional[str] = None) -> pd.DataFrame:
    """
    Build a multi-label dataframe for a list of glyphs.

    glyph_folder_map: optional dict mapping glyph identifier -> folder name where its images live.
                      If provided, full_image_path = os.path.join(base_image_root, folder, glyph_image).
                      Otherwise full_image_path = os.path.join(base_image_root, glyph_image).
    """
    df = pd.read_csv(glyph_elements_csv, dtype=str).fillna('')

    # filter by provided glyph list
    df_filtered = df[df[glyph_key].isin(glyph_list)].copy()
    if df_filtered.empty:
        raise ValueError("No rows matched the provided glyph_list and glyph_key.")

    # choose which column to parse for labels
    label_col = 'element_ids' if use_element_ids else 'elements'

    # parse pipe-separated labels into lists (keep non-empty tokens)
    def parse_pipe(s):
        if not s or pd.isna(s):
            return []
        return [tok.strip() for tok in s.split('|') if tok.strip()]

    label_lists = df_filtered[label_col].apply(parse_pipe).tolist()

    # build binary matrix using MultiLabelBinarizer
    mlb = MultiLabelBinarizer(sparse_output=False)
    if len(label_lists) == 0 or all(len(l) == 0 for l in label_lists):
        labels_df = pd.DataFrame(index=df_filtered.index)
    else:
        labels_mat = mlb.fit_transform(label_lists)
        if use_element_ids:
            col_names = [f"el_{c}" for c in mlb.classes_]
        else:
            col_names = [str(c).replace(' ', '_').replace('/', '_') for c in mlb.classes_]
        labels_df = pd.DataFrame(data=labels_mat, columns=col_names, index=df_filtered.index)

    # add full image path using optional per-glyph folder map
    def make_path(row):
        fn = row.get('glyph_image', '')
        if not fn:
            return ''
        glyph = row.get(glyph_key)
        if glyph_folder_map:
            folder = glyph_folder_map.get(glyph)
            if folder:
                return os.path.join(base_image_root, str(folder), fn)
        return os.path.join(base_image_root, fn)

    df_filtered['full_image_path'] = df_filtered.apply(make_path, axis=1)

    # assemble final dataframe
    final_df = pd.concat([df_filtered[[glyph_key, 'glyph_image', label_col, 'full_image_path']].rename(
                              columns={label_col: 'elements_original'}),
                         labels_df], axis=1)

    if save_csv:
        final_df.to_csv(save_csv, index=False)

    return final_df

# --- Example usage with dict mapping glyph -> folder name ---
glyphs_dict = {
    "032_05_003": "acatl-glyph",
    "026r_A_07": "atl-glyph",
    "032_05_005": "calli-glyph",
    "026r_B_02": "chimalli-glyph",
    "032_05_032": "cohuatl-glyph",
    "032_05_024": "cuauhtli-glyph",
    "027_05": "pantli-glyph",
    "115_11_28": "tetl-glyph",
    "032_05_012": "tochtli-glyph",
    "385_17r_17_02": "xiuguitzolli-glyph"
}

glyph_list = list(glyphs_dict.keys())

df_multi = make_multilabel_df(
    glyph_list=glyph_list,
    glyph_key='glyph_cote',
    glyph_elements_csv='glyph_elements.csv',
    base_image_root=r"f:\CODEX\AI_clinic\glyphs",   # root folder containing per-glyph folders
    glyph_folder_map=glyphs_dict,
    use_element_ids=False,
    save_csv='train_multilabel.csv'
)

print("Final shape:", df_multi.shape)
```

```python
# ...existing code...
import os
import glob
import pandas as pd

# root where per-glyph subfolders live
glyphs_root = r"f:\CODEX\AI_clinic\glyphs"

# find all JPG images under glyphs_root (recursively)
all_paths = glob.glob(os.path.join(glyphs_root, "**", "*.jpg"), recursive=True)
print(f"Found {len(all_paths)} image files under {glyphs_root}")

# local basename -> parent folder
local_parent = {os.path.basename(p): os.path.basename(os.path.dirname(p)) for p in all_paths}
local_basenames = list(local_parent.keys())

# read CSV and build lower->original lookup for glyph_image
df_g = pd.read_csv("glyph_elements.csv", dtype=str).fillna('')
csv_lookup = {name.lower(): name for name in df_g['glyph_image'].unique()}

# match local files to CSV (case-insensitive), build folder map keyed by CSV filename
matched = []
glyph_folder_map = {}
not_matched = []
for lb in local_basenames:
    key = lb.lower()
    if key in csv_lookup:
        csv_name = csv_lookup[key]
        matched.append(csv_name)
        glyph_folder_map[csv_name] = local_parent.get(lb, '')
    else:
        not_matched.append(lb)

print(f"Matches in glyph_elements.csv: {len(matched)} / {len(local_basenames)}")
if not_matched:
    print("Sample files not matched in CSV:", not_matched[:10])

# Use existing function to build multilabel DataFrame by glyph_image (pass CSV-original names)
df_multi = make_multilabel_df(
    glyph_list=matched,
    glyph_key='glyph_image',
    glyph_elements_csv='glyph_elements.csv',
    base_image_root=glyphs_root,
    glyph_folder_map=glyph_folder_map,
    use_element_ids=False,
    save_csv='train_multilabel_from_existing_images.csv'
)

print("Final shape:", df_multi.shape)
# ...existing code...
```
